```python
import glob
import numpy
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = sorted(glob.glob('data*.csv'))
dataset
```

```python
def visualize(dataset):
    average_list = []
    min_list = []
    max_list = []
    
    dataset_read = pd.read_csv(dataset)
    for columns in dataset_read:
        average_list.append(dataset_read[columns].mean())
        min_list.append(dataset_read[columns].min())
        max_list.append(dataset_read[columns].max())
     
    plt.figure(figsize=(10,7))
    plt.plot(average_list, label = 'Average')
    plt.plot(min_list, label = 'Minimum')
    plt.plot(max_list, label = 'Maximum')
    
    plt.title(dataset)
    plt.xlabel('Features')
    plt.ylabel('Values')
    plt.legend()
    plt.show()
```

```python
def identify_issue(dataset):
    dataset_read = pd.read_csv(dataset)
    min_list = []
    if dataset_read.iloc[:, 0].max() == 0 and dataset_read.iloc[:, 20].max() == 20:
        print('I doubt the integrity of the dataset')
    else:
        for columns in dataset_read:
            min_list.append(dataset_read[columns].min())
        min_sum = sum(min_list)
        if min_sum == 0:
            print('Why is this happening?')
        else:
            print('Dataset looks good.')
```

```python
for i in dataset:
    visualize(i)
    identify_issue(i)
```
